### Data Visualization
Data visualization is the graphical representation of data and information. It involves creating visual elements such as charts, graphs, and maps to help audiences understand the significance of data by presenting it in a visual context

This can be very helpful when discovering and getting to know a dataset and can help with classifying patterns, corrupt data, outliers, and much more. With a little domain knowledge, data visualizations can be used to express and demonstrate key relationships in plots and charts

### Importance of Data Visualisation
1.Exploratory Data Analysis (EDA)

2.Feature Engineering

3.Model Evaluation

4.Error Analysis

5.Communication

### Matplotlib(Python Plotting Library)
Human minds are more adaptive for the visual representation of data rather than textual data. We can easily understand things when they are visualized. It is better to represent the data through the graph where we can analyze the data more efficiently and make the specific decision according to data analysis. Before learning the matplotlib, we need to understand data visualization and why data visualization is important.

### Matplotlib
Matplotlib is a Python library which is defined as a multi-platform data visualization library built on Numpy array. It can be used in python scripts, shell, web application, and other graphical user interface toolkit.

#### Importance of Matplotlib
Certainly, here are the major importance of Matplotlib in a simple, point-wise format:

1. **Versatility:** Matplotlib offers a wide range of plot types and customization options.
  
2. **Effectiveness:** It helps create clear, informative, and visually appealing plots for data exploration and communication.
  
3. **Integration:** Matplotlib seamlessly integrates with other Python libraries and frameworks commonly used in data analysis.
  
4. **Publication-Quality Output:** It facilitates the generation of professional-looking plots suitable for academic papers, reports, and presentations.
  
5. **Community Support:** Matplotlib benefits from a large and active user community, providing extensive documentation and resources for users.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)#11 linearly space values between 0 to 5
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
#To Change Color of the graph
plt.plot(x,y,"r")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
#Dash lines
plt.plot(x,y,"r--")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
#To plot green line with asterisk marker and dash lines
plt.plot(x,y,"g*--")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
plt.scatter(x,y,color="red",marker="*")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

### Subplot

```python
#Plotting 2 graphs on a Single Figure
plt.subplot(2,2,1)#axis position
plt.plot(x,y,"g*--")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")

plt.subplot(2,2,4)
plt.scatter(x,y,color="green",marker = "*")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
plt.subplot(2,2,1)#axis position
plt.plot(x,y,"g*--")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")

plt.subplot(2,2,4)
plt.scatter(x,y,color="green",marker = "*")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")

plt.subplot(2,2,2)
plt.plot(x,y,"r--")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
#Bar graph
plt.bar(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
plt.bar(x,y,color = "green")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("My Title")
```

```python
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 9, 6]
plt.bar(x,y)
```

```python
#Pie Chart
labels = ["SDLC", "DM", "FLAT", "ML"]
number = [75,80,60,90]
plt.pie(number,labels = labels)
```

```python
labels = ["SDLC", "DM", "FLAT", "ML"]
number = [75,80,60,90]
plt.pie(number,labels = labels,colors = ["orange", "blue", "yellow","black"])
```

```python
labels = ["SDLC", "DM", "FLAT", "ML"]
number = [75,80,60,90]
plt.pie(number,labels = labels,autopct = "%0.1f%%",colors = ["orange", "blue", "yellow","black"])
```

```python
#Plotting all the graph in the single plot with different size

labels = ["SDLC", "DM", "FLAT", "ML"]
number = [75,80,60,90]
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
plt.pie(number,labels = labels,autopct = "%0.1f%%",colors = ["orange", "blue", "yellow","black"])
ax1.plot(x,y,"r--")
ax2.scatter(x,y,color = "green",marker = "*")
ax3.bar(x,y,color = "blue")
```

```python
#Plotting two lines in the same graph
plt.plot(x,y,"r--")
plt.plot(y,x,"g--")
```
