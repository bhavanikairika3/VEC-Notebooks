# Histogram
### A distribution plot, also known as a histogram, is a graphical representation of the frequency distribution of a dataset. It shows the distribution of values along an axis, typically the x-axis, and the frequency of occurrence of these values along the y-axis. In Seaborn, you can create a distribution plot using the sns.histplot() function.

```python
import matplotlib as plt
```

```python
sns.displot(data['total_bill'])
plt.title("Histogram")
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.displot(data["total_bill"], color="green")
```

# Histogram using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
plt.hist(data['total_bill'])
plt.title("Histogram using Plt")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
```

```python
sns.displot(data.total_bill,kind="kde")
plt.title("Kde plot")
plt.xlabel("Total Bill")
plt.ylabel("Density")
```

```python
sns.displot(data.total_bill,kind="kde", color="green")
plt.title("Kde plot")
plt.xlabel("Total Bill")
plt.ylabel("Density")
```

# Scatter Plot
### A scatter plot is a diagram where each value in the data set is represented by a dot. The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, one for the values of the x-axis, and one for the values of the y-axis

```python
sns.scatterplot(data=data, x="total_bill", y="tip")
```

# Joint Plot
### A Jointplot is a figure that showcases the relationship between two variables, combining scatter plots, hexbin plots, regression plots, or 2D kernel density plots with histograms, KDE (Kernel Density Estimate) plots, or other representations of the univariate distribution of each variable on the margins.

```python
sns.jointplot(data=data, x="total_bill", y="tip")
```

```python
sns.jointplot(data=data, x="total_bill", y="tip", kind="kde")
```

# Box Plot
### Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

```python
sns.boxplot(data.total_bill)
```

# Heatmap
### A heatmap (or heat map) is a graphical representation of data where values are depicted by color. They are essential in detecting what does or doesn't work on a website or page, and which parts and elements of a page users engage with.


```python
import pandas as pd
data1=pd.get_dummies(data)
a = data1.corr()
```

```python
sns.heatmap(a)
```

```python
sns.heatmap(a,annot=True)
```

```python
print(a)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(a, annot=True, fmt=".2f", cmap="coolwarm", linewidth=0.5, annot_kws={"size":10})
```

# Lineplot
### Line graphs are used to track changes over short and long periods of time. When smaller changes exist, line graphs are better to use than bar graphs.

```python
sns.relplot(data=data, x = "total_bill", y = "tip", kind = "line")
```
